```python
import numpy as np
from matplotlib import pylab as plt
```

```python
def norm_pdf(x, mu=0, sigma=5):
    return 1./(sigma * np.sqrt(2 * np.pi)) * np.exp(-1./2 * ((x-mu)/sigma**2)**2)

def uniform_pdf(x_list, a=-2.5, b=2.5):
    ret = np.copy(x_list)
    for i, x in enumerate(x_list):
        if (a <= x) and (x <= b):
            ret[i] = 1./ (b-a)
        else:
            ret[i] = 0

    return ret

def f(x, t, func = norm_pdf):
    return t * func(t * x)
```

```python
x = np.arange(-3, 3, 0.001)
_, ax = plt.subplots(1, 2, figsize=(25, 12))
colors = ['red', 'green', 'blue', 'purple',
          'brown', 'orange', 'pink']

ts_normal = [1, 10, 50, 100, 200, 300]
ts_uniform = [1, 5, 10, 15, 25, 40]
size = 25

for t, color in zip(ts_normal, colors):
    ax[0].plot(x, f(x, t=t, func=norm_pdf), color=color, 
               label=f"t = {t}", linewidth=2)
    #ax[0].set_title(r"Normal distribution $N(0, 1)$", fontsize=15)
    ax[0].set_xlabel(r"$x$", fontsize=size)
    ax[0].set_ylabel(r"$f_t(x) = t \cdot f_0(tx)$", fontsize=size)

for t, color in zip(ts_uniform, colors):
    ax[1].plot(x, f(x, t=t, func=uniform_pdf), color=color, 
               label=f"t = {t}", linewidth=2)
    #ax[1].set_title(r"Uniform distribution $U[0, 1]$", fontsize=15)
    ax[1].set_xlabel(r"$x$", fontsize=size)
    ax[1].set_ylabel(r"$f_t(x) = t \cdot f_0(tx)$", fontsize=size)

ax[0].grid()
ax[1].grid()
ax[0].legend(fontsize=size)
ax[1].legend(fontsize=size)

plt.show()
```
